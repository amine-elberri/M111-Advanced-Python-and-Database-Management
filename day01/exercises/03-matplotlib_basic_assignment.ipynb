## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="https://github.com/amine-elberri/M111-Advanced-Python-and-Database-Management/blob/main/day01/data/Company_profit_per_month.png?raw=1" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(10, 5))
url ='https://raw.githubusercontent.com/amine-elberri/M111-Advanced-Python-and-Database-Management/main/day01/data/company_sales_data.csv'
data  = pd.read_csv(url,delimiter=",")
data
plt.plot(data['month_number'],data['total_profit'])
plt.ylabel('Profits in dollar')
plt.xlabel("Month number")
plt.title("Monthly profit")
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="https://github.com/amine-elberri/M111-Advanced-Python-and-Database-Management/blob/main/day01/data/sales_data_pi_chart.png?raw=1" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
plt.figure(figsize=(20, 10))

sales_by_product = data.groupby("month_number").tail(12)[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]].sum()
plt.pie(sales_by_product, labels=sales_by_product.index, autopct='%1.1f%%')
plt.title("Sales data")
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="https://github.com/amine-elberri/M111-Advanced-Python-and-Database-Management/blob/main/day01/data/Sales_data_of_a_facewash.png?raw=1" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
plt.figure(figsize=(15, 6))
soap_sales = data[data["bathingsoap"] > 0]
facewash_sales = data[data["facewash"] > 0]

soap_months = soap_sales["month_number"]
soap_sales = soap_sales["bathingsoap"]
facewash_months = facewash_sales["month_number"]
facewash_sales = facewash_sales["facewash"]
fig, axs = plt.subplots(2, 1)

axs[0].plot(soap_months, soap_sales,color='black', marker='o', linewidth=3)
axs[0].set_xlabel("Month")
axs[0].set_ylabel("Sales units in number")
axs[0].set_title("Sales data of Bathing Soap")

axs[1].plot(facewash_months, facewash_sales,color='red',marker='o', linewidth=3)
axs[1].set_xlabel("Month")
axs[1].set_ylabel("Sales units in number")
axs[1].set_title("Sales data of Facewash")

plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="https://github.com/amine-elberri/M111-Advanced-Python-and-Database-Management/blob/main/day01/data/product_sales_stack_plot.png?raw=1" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
plt.figure(figsize=(15, 6))


sales_by_month = data.groupby("month_number")[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]].sum()


months = sales_by_month.index
facecream_sales = sales_by_month["facecream"]
facewash_sales = sales_by_month["facewash"]
toothpaste_sales = sales_by_month["toothpaste"]
bathingsoap_sales = sales_by_month["bathingsoap"]
shampoo_sales = sales_by_month["shampoo"]
moisturizer_sales = sales_by_month["moisturizer"]


plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales, labels=["Face Cream", "Facewash", "Toothpaste", "Bathing Soap", "Shampoo", "Moisturizer"],colors=['m','c','r','k','g','y'])
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title("All products sales data using stack plot")
plt.legend(loc="upper left")


plt.show()
```
